### Поиск начала цикла ходьбы

### Тут надо пояснить


Берем некоторое окно и смотрим производную, геоментрическую производную, при этом интерполируем и рассматриваем определенный участок

```python
start_frame_number = 0 #номер фрейма с которого начинается
col = 'rfoot_coord_1' #название столбца
df = pd.DataFrame(d[col])
d_final = d.copy(deep=True)
win = 15
thr = 0.4
```

```python
for i in range(1, win + 1):
    df['diff_' + str(i)] = df[col].diff(i)
cols = df.columns.tolist()
cols.remove(col)
df['med_diff'] = df[cols].median(axis=1)
df['monoton'] = 0
df.loc[df['med_diff'] > thr, 'monoton'] = 1

plt.plot(d['frame'], df['monoton'])
plt.plot(d['frame'], d['rfoot_coord_1'])
#надо найти эти самые точки перегиба(локальные минимумы)
```

```python
positions = []
list_values = df['monoton'].tolist()
for ix in range(len(list_values)):
    if list_values[ix] > 0 and list_values[ix+1] == 0:
        positions.append(ix)
        print ('Max point ',ix)
```

Далее необходимо избавиться от проблемы, что в начале может быть половина цикла, надо это убрать<br />
посколько расстояние между ними разное, то берем за начало и конец те, расстояние между которыми больше

```python
minimum_pos = []
start = 0
end = 0
# minimum = max(df['lfoot_coord_1'][start:end].tolist())
df['rfoot_coord_1'][start:end].tolist()

for ix in range (len(positions) - 2):
    if positions[ix + 1] - positions[ix] < positions[ix + 2] - positions[ix + 1]:
        #если расстояние между текущим и следующим меньше, чем расстоние между следующим и через 2 впередеи
        #то два соседних горба попали в начало, значит отчет горбов начался корректно
        start = positions[ix]
        end = positions[ix + 1]
        #сохраним точку минимума между соседними горбами
        minimum = max(df['rfoot_coord_1'][start:end].tolist())
        for ixn in range(start,end):
            
            per = float(df['rfoot_coord_1'][ixn - 1:ixn].tolist()[0])
            if per < minimum:
                minimum = float(df['rfoot_coord_1'][ixn - 1:ixn].tolist()[0])
                minimum_p = ixn
        minimum_pos.append(minimum_p)
        print('++', ix)
    else:
        start = positions[ix + 1]
        end = positions[ix + 2]
        minimum = max(df['rfoot_coord_1'][start:end].tolist())
        for ixn in range(start,end):
            per = float(df['rfoot_coord_1'][ixn - 1:ixn].tolist()[0])
            if per < minimum:
                minimum = float(df['rfoot_coord_1'][ixn - 1:ixn].tolist()[0])
                minimum_p = ixn
        minimum_pos.append(minimum_p)
        print('---', ix)
        
#delete dublicates
minimum_pos = list( dict.fromkeys(minimum_pos) ) 
print(minimum_pos)
```
