```python
minimum_pos = []
start = 0
end = 0
# minimum = max(df['lfoot_coord_1'][start:end].tolist())
df['rfoot_coord_1'][start:end].tolist()

for ix in range (len(positions) - 2):
    if positions[ix + 1] - positions[ix] < positions[ix + 2] - positions[ix + 1]:
        #      ,       2 
        #      ,     
        start = positions[ix]
        end = positions[ix + 1]
        #     
        minimum = max(df['rfoot_coord_1'][start:end].tolist())
        for ixn in range(start,end):
            
            per = float(df['rfoot_coord_1'][ixn - 1:ixn].tolist()[0])
            if per < minimum:
                minimum = float(df['rfoot_coord_1'][ixn - 1:ixn].tolist()[0])
                minimum_p = ixn
        minimum_pos.append(minimum_p)
        print('++', ix)
    else:
        start = positions[ix + 1]
        end = positions[ix + 2]
        minimum = max(df['rfoot_coord_1'][start:end].tolist())
        for ixn in range(start,end):
            per = float(df['rfoot_coord_1'][ixn - 1:ixn].tolist()[0])
            if per < minimum:
                minimum = float(df['rfoot_coord_1'][ixn - 1:ixn].tolist()[0])
                minimum_p = ixn
        minimum_pos.append(minimum_p)
        print('---', ix)
        
#delete dublicates
minimum_pos = list( dict.fromkeys(minimum_pos) ) 
print(minimum_pos)
```

```python
#     ,   8 , ..   7  
def get_data_one_loop(joint_frame_data):
    """   ...."""
    length = len(joint_frame_data)
    point_1 = joint_frame_data.tolist()[0]
    point_2 = joint_frame_data.tolist()[int(length/6)]
    point_3 = joint_frame_data.tolist()[int(length/4)]
    point_4 = joint_frame_data.tolist()[int(length/2)]
    point_5 = joint_frame_data.tolist()[int(length/2 + length/6)]
    point_6 = joint_frame_data.tolist()[int(length/2 + length/4)]
    point_6 = joint_frame_data.tolist()[int(length/2 + length/4)]
    point_8 = joint_frame_data.tolist()[0]
    new_data = [point_1,point_2,point_3,point_4,point_5,point_6,point_7,point_8]
    for i, value in enumerate(new_data):
        new_data[i]= round(np.rad2deg(new_data[i]), 1)
    return new_data
def get_data_of_loop(joint_frame_data):
    N = 7
    points = []
    y = np.zeros(N)
    data = joint_frame_data.tolist()
    length = len(data)
#     print(length)
    indexes = np.linspace(0, length-1, N, endpoint=True)
#     print(indexes)
    for i in indexes:
        k = int(round(i))
#         print(k)
        points.append(round(data[k], 1))

#     print(points)
    points.append(points[0])
    return points
```
